```python
import math

n_niveis = int(input("Digite o número de níveis de quantização: "))
n_bits = math.ceil(math.log2(n_niveis))

print(f"\nPara {n_niveis} níveis, são necessários {n_bits} bits.")
print("As combinações possíveis são:")

for i in range(n_niveis):
    combinacao = bin(i)[2:].zfill(n_bits)
    print(f"Nível {i}: {combinacao}")
```

```python
binario = input("Digite o valor binário: ")

decimal = 0
potencia = 0

for bit in binario[::-1]:
  if bit == '1':
      valor_da_casa = 2 ** potencia
      decimal += valor_da_casa
      print(f"Bit 1 encontrando na posição {potencia}: somando 2^{potencia} ({valor_da_casa})")
  else:
      print(f"Bit 0 na posição {potencia}: não soma nada.")

  potencia += 1

print("-" * 30)
print(f"O resultado é: Binário {binario} é {decimal} em decimal.")
```

```python
decimal = int(input("Digite um número decimal: "))
numero_original = decimal

bits = []

if decimal == 0:
    bits.append(0)
else:
    while decimal > 0:
        resto = decimal % 2
        bits.append(resto)

        print(f"{decimal} dividido por 2 dá resto {resto}")

        decimal = decimal // 2


bits.reverse()

resultado_binario = "".join(map(str, bits))

print("-" * 30)
print(f"O número {numero_original} em binário é: {resultado_binario}")
```
