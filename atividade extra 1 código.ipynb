```python
import math

# 1. Entrada do usuário
n_niveis = int(input("Digite o número de níveis de quantização: "))

# 2. Cálculo do número de bits
# Usamos o logaritmo na base 2 e arredondamos para cima
n_bits = math.ceil(math.log2(n_niveis))

print(f"\nPara {n_niveis} níveis, são necessários {n_bits} bits.")
print("As combinações possíveis são:")

# 3. Gerar as combinações
for i in range(n_niveis):
    # Converte o número para binário e remove o prefixo '0b'
    # O 'zfill' garante que todos tenham o mesmo número de bits (preenche com zeros)
    combinacao = bin(i)[2:].zfill(n_bits)
    print(f"Nível {i}: {combinacao}")
```
